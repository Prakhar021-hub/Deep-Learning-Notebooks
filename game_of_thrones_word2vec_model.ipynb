```python
import os
import nltk
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import zipfile
import numpy as np
import pandas as pd

# Download punkt data
nltk.download("punkt")
nltk.download("punkt_tab")

# 1. Download dataset
!kaggle datasets download -d khulasasndh/game-of-thrones-books

# 2. Unzip into a folder
with zipfile.ZipFile("game-of-thrones-books.zip", "r") as zip_ref:
    zip_ref.extractall("got_books")

# 3. Read all .txt files
story = []

for filename in os.listdir("got_books"):
    if filename.endswith(".txt"):   # only text files
        with open(os.path.join("got_books", filename), encoding="utf-8", errors="ignore") as f:
            corpus = f.read()
            raw_sent = sent_tokenize(corpus)   # sentence split
            for sent in raw_sent:
                story.append(simple_preprocess(sent))  # tokenize words

```

```python
story
```

```python
from gensim.models import Word2Vec

# Train Word2Vec model
model = Word2Vec(
    sentences=story,   # our tokenized sentences
    vector_size=100,   # size of word vectors (dimensions)
    window=10,         # context window
    min_count=2,       # ignore rare words
    workers=4,         # use 4 CPU cores
    sg=0               # 0 = CBOW, 1 = Skip-Gram
)

# Save model (optional)
model.save("got_word2vec.model")
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
