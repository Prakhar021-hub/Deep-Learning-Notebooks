```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data["input1"] = [1,1,0,0]
or_data["input2"] = [1,0,1,0]
or_data["output"] = [1,1,1,0]
or_data
```

```python
and_data["input1"] = [1,1,0,0]
and_data["input2"] = [1,0,1,0]
and_data["output"] = [1,0,0,0]
and_data
```

```python
xor_data["input1"] = [1,1,0,0]
xor_data["input2"] = [1,0,1,0]
xor_data["output"] = [1,0,0,1]
xor_data
```

```python
sns.scatterplot(x = and_data["input1"], y = and_data["input2"] , hue = and_data["output"])
```

```python
sns.scatterplot(x = or_data["input1"], y = or_data["input2"] , hue = or_data["output"])
```

```python
sns.scatterplot(x = xor_data["input1"], y = xor_data["input2"] , hue = xor_data["output"])
```

```python
from sklearn.linear_model import Perceptron
```

```python
classifier_and = Perceptron()
classifier_or = Perceptron()
classifier_xor = Perceptron()
```

```python
classifier_and.fit(X = and_data.iloc[:, 0:2].values , y = and_data.iloc[:, -1].values)
classifier_or.fit(X = or_data.iloc[:, 0:2].values , y = or_data.iloc[:, -1].values )
classifier_xor.fit(X = xor_data.iloc[:, 0:2].values , y = xor_data.iloc[:, -1].values)
```

```python
classifier_and.intercept_
```

```python
classifier_and.coef_
```

```python
from mlxtend.plotting import plot_decision_regions

fig, axes = plt.subplots(1,3, figsize = (18,5))


## AND gate Decision Boundary
axes[0].set_title("And Gate Decision Boundary")
plot_decision_regions(X = and_data.iloc[:, 0:2].values , y = and_data.iloc[:, -1].values, clf = classifier_and, legend = 2, ax= axes[0])

## OR gate Decision Boundary
axes[1].set_title("OR Gate Decision Boundary")
plot_decision_regions(X = or_data.iloc[:, 0:2].values , y = or_data.iloc[:, -1].values, clf = classifier_or, legend = 2, ax = axes[1])

## XOR gate Decision Boundary
axes[2].set_title("XOR Gate Decision Boundary")
plot_decision_regions(X = xor_data.iloc[:, 0:2].values , y = xor_data.iloc[:, -1].values, clf = classifier_xor, legend = 2 , ax =axes[2])

plt.show()
```
